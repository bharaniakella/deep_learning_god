```python
#log is a monotonous function -> logx increases as x increases. logx decreases as x decreases
#logx helps in distinguishing extremely small values like probabilities.
#natural logx is inverse of exp(x). Both cancel out each other
```

```python
import numpy as np
from matplotlib import pyplot as plt

x=np.linspace(0.001,1,150)

logx=np.log(x)

plt.plot(x,logx,'o-')
plt.show()
```

```python
expx=np.exp(x)

plt.plot(x,expx,'o-')
plt.show()
```

```python
plt.plot(x,x)
plt.plot(x,expx,'x')
plt.plot(x,logx,'o-')
plt.legend(['x vs x','x vs exp x','x vs logx'])
plt.show()
```

```python
plt.plot(x,np.log(expx))
plt.legend(['x vs log(exp(x))'])
plt.show()
```

```python
plt.plot(x,np.log(expx),color="green")
plt.legend(['x vs exp(log(x))'])
plt.show()
```
